```r
# nrow(gwas_annos)
# colnames(gwas_annos)
mycols = c("feature","snp_id","stage","beta.mean","p_value.mean","global_corr_p_value","stage_specific",
           "snp_chromosome","snp_position","ref_allele","alt_allele",
           "p_value.int","q_value","beta.int","sign.int",
           "GWAS_Variant","GWAS_Annotation","GWAS_LD")
head(gwas_annos[,mycols])
```

```r
sc_qtl_filename = "/hps/nobackup/hipsci/scratch/singlecell_endodiff/results_files_20181031/SuppTable_S5_singlecell_eQTLs.tsv"
sc_qtl = read.csv(sc_qtl_filename, sep = "\t")
head(sc_qtl)
```

```r
gwas_annos$ensembl_gene_id = gsub("_.*","",gwas_annos$feature)
head(gwas_annos$ensembl_gene_id)
```

```r
sc_qtl_gwas = left_join(sc_qtl,gwas_annos[,c("ensembl_gene_id","snp_id","stage",
                                              "GWAS_Annotation")])
```

```r
nrow(sc_qtl_gwas)
head(sc_qtl_gwas)
```

```r
sc_qtl_gwas$GWAS_tagging = (sc_qtl_gwas$GWAS_Annotation != "-")
head(sc_qtl_gwas)
```

```r
sc_qtl_gwas$GWAS_Annotation <- c()
head(sc_qtl_gwas)
```

```r
write.table(sc_qtl_gwas,
            "/hps/nobackup/hipsci/scratch/singlecell_endodiff/results_files_20181031/SuppTable_S3_singlecell_eQTLs.tsv",
           quote = F, row.names = F)
```
