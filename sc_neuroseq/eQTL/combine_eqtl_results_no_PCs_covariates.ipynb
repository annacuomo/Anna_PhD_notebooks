```r
annos_file = "/hps/nobackup/stegle/users/acuomo/all_scripts/Ensemble_75_Gene_Structures_Exon_Info.txt"
annos = read.csv(annos_file, sep = "\t")
head(annos,2)
```

```r
annos1 = distinct(annos[,c("Ensembl.Gene.ID","Gene.Start..bp.","Gene.End..bp.","Strand")])
colnames(annos1) = c("feature_id","feature_start","feature_end","strand")
head(annos1,2)
```

```r
nrow(myfile2)
myfile3 = left_join(myfile2,annos1)
nrow(myfile3)
```

```r
# reference
colnames(myfile3)
myfile3$dist_from_tss = 0
for (i in 1:nrow(myfile3)){
    snp = myfile3$snp_position[i]
    start = myfile3$feature_start[i]
    end = myfile3$feature_end[i]
    strand = myfile3$strand[i]
    if (strand==1){
        TSS = start
        myfile3$dist_from_tss[i] = snp-TSS 
    }
    else if (strand==-1){
        TSS = end
        myfile3$dist_from_tss[i] = TSS-snp
    }
}
head(myfile3)
```
