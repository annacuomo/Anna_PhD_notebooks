```r
eqtl_in_tissue = function(leads, results, tissue, T = 0.05){
    res = inner_join(leads, results, by = c("snp_id","feature_id"), suffix = c(".leads",".results"))
    # sharing defined as nominal significance (T) and same direction of effect
    res[[tissue]] = 0
    res[(res$p_value.results > T),tissue] = 1
    res
}
```

```r
matrix_corr = function(list_leads, list_results){
    mat = matrix(0, nrow = length(list_leads), ncol = length(list_results))
    for (i in 1:length(list_leads)){
#         print(i)
        leads = list_leads[[i]]
        for (j in 1:length(list_results)){
            results = list_results[[j]]
            tmp = inner_join(leads, results, by = c("snp_id","feature_id"), suffix = c(".leads",".results"))
            mat[i,j] = cor(tmp$beta.leads, tmp$beta.results)
        }
            
    }
    mat
}
```
