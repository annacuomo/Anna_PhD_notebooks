```r
for (j in 1:length(celltypes)){
# for (j in 1){
    celltype = as.character(celltypes[j])
#     print(celltype)
    egs = egenes[[celltype]]
    for (gene in egs){
        n = 0
#         print(gene)
        for (i in 1:length(brain_tissues)){
            ct = brain_tissues[i]
            print(ct)
            if (gene %in% egenes[[ct]]){
                n = n+1
            }
        }
        mat[i,j] = n
    }
}
```

```r
for (j in 1:length(celltypes)){
# for (j in 1){
    celltype = as.character(celltypes[j])
#     print(celltype)
    egs = egenes[[celltype]]
    
    for (i in 1:length(brain_tissues)){
        ct = brain_tissues[i]
#         print(ct)
        mat[i,j] = length(egs[!(egs %in% egenes[[ct]])])/length(egs)
    }
}
```
