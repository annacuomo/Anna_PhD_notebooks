```r
gene = 'ENSG00000211643'
```

```r
## get LD matrix
R_file = paste0(LD_folder,gene,".csv")
R = read.csv(R_file, row.names=1)
colnames(R) = gsub("X","",colnames(R))
head(R)
```

```r
R2 = R
dim(R2)
```

```r
snps1 = unique(as.character(rownames(R1)))
snps2 = unique(as.character(rownames(R2)))
common_snps = snps1[snps1 %in% snps2]
```

```r
RR1 = R1[common_snps, common_snps]
RR1 = RR1[order(rownames(RR1)),order(colnames(RR1))]
```

```r
RR2 = R2[common_snps, common_snps]
RR2 = RR2[order(rownames(RR2)),order(colnames(RR2))]
```

```r
head(RR1)
head(RR2)
```

```r
# all results
myfile0 = paste0(revision_folder,"Marcs_results_all.csv")
df0 = read.csv(myfile0, row.names=1)
```

```r
## get Z scores
df_rel = df0[df0$feature_id == gene,]
df_rel$z = df_rel$beta / df_rel$beta_se
nrow(df_rel)
length(unique(df_rel$snp_id))
head(df_rel)
```

```r
snps3 = unique(as.character(df_rel$snp_id))
common_snps2 = snps3[snps3 %in% snps2]
```

```r
df_rel = df_rel[df_rel$snp_id %in% common_snps,]
df_rel = df_rel[df_rel$snp_id %in% common_snps2,]
df_rel = df_rel[order(df_rel$snp_id),]
```

```r
z_scores = df_rel$z
```

```r
RR3 = R2[common_snps2, common_snps2]
RR2 = RR3[order(rownames(RR3)),order(colnames(RR3))]
# run Susie using summary stats
fitted_rss <- susie_rss(z_scores, as.matrix(RR3), L = 10)
susie_plot(fitted_rss, y="PIP")
```

```r
# run Susie using summary stats
fitted_rss <- susie_rss(z_scores, as.matrix(RR1), L = 10)
susie_plot(fitted_rss, y="PIP")
```

```r
# run Susie using summary stats
fitted_rss <- susie_rss(z_scores, as.matrix(RR2), L = 10)
susie_plot(fitted_rss, y="PIP")
```
