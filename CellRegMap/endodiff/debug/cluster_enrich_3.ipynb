```r
gene_list_list = list()
```

```r
for (i in 1:ncol(df)){
    gene_list = df[,i]
    names(gene_list) = rownames(df)
    # get positively correlated genes
    genes_pos = gene_list[gene_list>0.5]
#     if (length(genes_pos)==0){next}
    # order
    genes_pos = genes_pos[order(genes_pos, decreasing = T)]
    # get gene names only
    genes = gsub(".*_","",names(genes_pos))
    gene_list_list[[i]] = genes
}
```

```r
gres = gost(query=gene_list_list, organism = "hsapiens", ordered_query = TRUE, significant=TRUE, multi_query=TRUE, user_threshold = 5e-8)
```
