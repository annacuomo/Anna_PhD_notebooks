```python
import pandas as pd
import harmonypy as hm
```

```python
mydir = "/hps/nobackup/stegle/users/acuomo/all_scripts/struct_LMM2/sc_endodiff/"
```

```python
meta_data = pd.read_csv(mydir+"coldata_sce.csv", index_col=0)
meta_data.head()
```

```python
meta_data.shape
```

```python
vars_use = ['experiment']
```

```python
# data_mat = pd.read_pickle(mydir+"exprs_alldays_allgenes.tsv.pkl")
data_mat = pd.read_csv(mydir+"pcs_500hvgs.csv", index_col=0)
data_mat.head()
```

```python
data_mat.shape
```

```python
ho = hm.run_harmony(data_mat, meta_data, vars_use)
```

```python
res = pd.DataFrame(ho.Z_corr)
```

```python
res.shape
```

```python
res = res.T
```

```python
res.columns = ['hPC{}'.format(i + 1) for i in range(res.shape[1])]
res.index = data_mat.index
```

```python
res.head()
```

```python
# Write the adjusted PCs to a new file.
res.to_csv(mydir+"harmony_pcs.csv")
```
