```r
sce = scater::normalize(sce)
design = model.matrix(~ experiment+n_total_reads, data = colData(sce))
alt.fit = scran::trendVar(sce, design = design, use.spikes = FALSE)
alt.decomp = scran::decomposeVar(sce, alt.fit)
```

```r
head(alt.decomp)
```

```r
hvgs_df = alt.decomp[alt.decomp$FDR < 0.01,]
head(hvgs_df[order(hvgs_df$bio, decreasing = T),])
hvgs = rownames(hvgs_df[order(hvgs_df$bio, decreasing = T),])
rowData(sce)$is_hvg <- (rownames(sce) %in% hvgs)
```

```r
length(hvgs)
head(hvgs,5)
```

```r
proxy_genes = hvgs[1]
```

```r
for (gene1 in hvgs){
#     print(gene1)
    for (gene2 in proxy_genes){
#         print(gene2)
        g1 = logcounts(sce)[gene1,]
        g2 = logcounts(sce)[gene2,]
        r = cor(g1,g2)
#         print(r**2)
        if(r**2 <= 0.2){add <- "yes"}
        else{add <- "no"
            break}
#         print(add)
    }
    if (add == "yes" & !(gene1 %in% proxy_genes)){
        proxy_genes <- c(proxy_genes, gene1)
    }
}
```

```r
for (gene1 in hvgs[1:2000]){
#     print(gene1)
    for (gene2 in proxy_genes){
#         print(gene2)
        g1 = logcounts(sce)[gene1,]
        g2 = logcounts(sce)[gene2,]
        r = cor(g1,g2)
#         print(r**2)
        if(r**2 > 0.2){break}
        else{
            proxy_genes <- unique(c(proxy_genes, gene1))
        }
    }
}
```
