```r
options(repr.plot.width = 5, repr.plot.height = 1.5)
m=min(df0$gene)
M=max(df0$gene)
ggplot(df0, aes (x = MOFA1, y = gene, col=gene)) + geom_point(alpha=0.2) + stat_smooth(method="gam") + theme_minimal() + ylim(c(m,M)) + scale_colour_gradientn(colors = rev(brewer.pal(9,"Spectral")))
ggplot(df0, aes (x = MOFA2, y = gene, col=gene)) + geom_point(alpha=0.2) + stat_smooth(method="gam") + theme_minimal() + ylim(c(m,M)) + scale_colour_gradientn(colors = rev(brewer.pal(9,"Spectral")))
```

```r
genes = colnames(df)
options(repr.plot.width = 5, repr.plot.height = 1.5)
for (gene in genes){
    df0 = cbind(df_pcs[,c("MOFA1","MOFA2")], data.frame(gene = df[,gene]))
    m=min(df0$gene)
    M=max(df0$gene)
    p1=ggplot(df0, aes (x = MOFA1, y = gene)) + stat_smooth(method="loess") + theme_minimal() + ylim(c(m,M))
    p2=ggplot(df0, aes (x = MOFA2, y = gene)) + stat_smooth(method="loess") + theme_minimal() + ylim(c(m,M))
    print(plot_grid(p1,p2,ncol=2))
}
```
