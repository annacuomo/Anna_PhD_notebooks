```r
mydir = "/hps/nobackup/stegle/users/acuomo/all_scripts/struct_LMM2/sc_neuroseq/coloc_examples/"
myfiles = list.files(mydir)
myfiles
```

```r
## load individual eQTL genetic effects
df_gc = read.csv("/hps/nobackup/stegle/users/acuomo/all_scripts/struct_LMM2/sc_neuroseq/May2021/genetic_effect/MOFA10/eqtl_tot_genetic_effect_all.txt", sep="\t",row.names = 1)
colnames(df_gc) = gsub("_-","-",gsub("\\.","-",colnames(df_gc)))
head(df_gc,2)
nrow(df_gc)
```

```r
genes = as.character(df_eqtl$gene)
```

```r
col_day30 = canva_pal("Cheerful brights")(4)[1]
col_day52 = canva_pal("Cheerful brights")(4)[3]
col_day52_rot = canva_pal("Cheerful brights")(4)[2]
```

```r
options(repr.plot.width=8, repr.plot.height=3)
for (gene in genes){
    snp = df_eqtl[df_eqtl$gene == gene,"snp_id"]
    filename = paste0(mydir, myfiles[grep(gene, myfiles)])
    df_g = read.csv(filename)
    colnames(df_g)[1] = "donor"    
    df_e = data.frame(pseudocell = colnames(df0), Expression = df0[gene,])
    df_e$donor = gsub("--.*","",df_e$pseudocell)
    df = inner_join(df_e, df_g, by="donor")
    M = max(df$Expression)
    ##
    p1 = ggplot(df[grep("d30",df$pseudocell),],aes(x = as.factor(Genotype),y=Expression)) 
    p1 = p1 + xlab(snp) + ylab(gene) + geom_violin(col=col_day30) + theme_classic() + ylim(c(0,M))
    p1 = p1 + geom_quasirandom(size=0.5, alpha=0.2, col=col_day30) + geom_boxplot(width=0.1, col=col_day30)
    ##
    p2 = ggplot(df[grep("d52_untr",df$pseudocell),],aes(x = as.factor(Genotype),y=Expression)) 
    p2 = p2 + xlab(snp) + ylab(gene) + geom_violin(col=col_day52) + theme_classic() + ylim(c(0,M))
    p2 = p2 + geom_quasirandom(size=0.5, alpha=0.2, col=col_day52) + geom_boxplot(width=0.1, col=col_day52)
    p2 = p2 + theme(axis.text.x = element_text(size = 12),axis.text.y = element_text(size = 12))
    ##
    p3 = ggplot(df[grep("d52_tr",df$pseudocell),],aes(x = as.factor(Genotype),y=Expression)) 
    p3 = p3 + xlab(snp) + ylab(gene) + geom_violin(col=col_day52_rot) + theme_classic() + ylim(c(0,M))
    p3 = p3 + geom_quasirandom(size=0.5, alpha=0.2, col=col_day52_rot) + geom_boxplot(width=0.1, col=col_day52_rot)
    print(plot_grid(p1,p2,p3,ncol=3)) 
    
}
```

```r
options(repr.plot.width=5, repr.plot.height=3)
for (gene in genes){
    snp = df_eqtl[df_eqtl$gene == gene,"snp_id"]
    condition = df_eqtl[df_eqtl$gene == gene,"condition"]
    filename = paste0(mydir, myfiles[grep(gene, myfiles)])
    df_g = read.csv(filename)
    colnames(df_g)[1] = "donor"    
    df_e = data.frame(pseudocell = colnames(df0), Expression = df0[gene,])
    df_e$donor = gsub("--.*","",df_e$pseudocell)
    df_ge = inner_join(df_e, df_g, by="donor")
    df = inner_join(df_ge, data.frame(pseudocell = rownames(df_gc), beta = df_gc[,grep(gene,colnames(df_gc))]), by = "pseudocell")
    df = df[grep(condition, df$pseudocell),]
    M = max(df$Expression)
    qts = quantile(df$beta, probs = c(0.3,0.7))
    df1 = df[df$beta>as.numeric(qts[2]),]
    df2 = df[df$beta<as.numeric(qts[1]),]
    p1 = ggplot(df1,aes(x = as.factor(Genotype),y=Expression)) + theme_classic() + ggtitle("top")
    p1 = p1 + xlab(snp) + ylab(gene) + geom_violin(col="firebrick") + ylim(c(0,M))
    p1 = p1 + geom_quasirandom(size=0.5, alpha=0.2, col="firebrick") + geom_boxplot(width=0.1, col="firebrick")
    p1 = p1 + theme(axis.text.x = element_text(size = 12),axis.text.y = element_text(size = 12))
    p2 = ggplot(df2,aes(x = as.factor(Genotype),y=Expression)) + theme_classic() + ggtitle("bottom")
    p2 = p2 + xlab(snp) + ylab(gene) + geom_violin(col="cornflowerblue") + ylim(c(0,M))
    p2 = p2 + geom_quasirandom(size=0.5, alpha=0.2, col="cornflowerblue") + geom_boxplot(width=0.1, col="cornflowerblue")
    p2 = p2 + theme(axis.text.x = element_text(size = 12),axis.text.y = element_text(size = 12))
    print(plot_grid(p1,p2,ncol=2))    
}
```
