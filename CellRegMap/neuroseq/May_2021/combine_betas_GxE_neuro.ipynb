```r
mydir = "/hps/nobackup/stegle/users/acuomo/all_scripts/struct_LMM2/sc_neuroseq/May2021/genetic_effect/MOFA10/"
```

```r
mygenes = list.files(mydir,pattern = "beta_GxE.tsv")
length(mygenes)
head(mygenes)
```

```r
gene = mygenes[1]
df0 = read.csv(paste0(mydir,gene),sep='\t', row.names = 1)
# head(df0)
cells = rownames(df0)
length(cells)
```

```r
df = data.frame(cells = cells)
for (gene in mygenes){
#     print(gene)
    df_curr = read.csv(paste0(mydir,gene),sep='\t', row.names = 1)
    colnames(df_curr) = gsub("X._","",colnames(df_curr))
    colnames(df_curr) = gsub("X.._","",colnames(df_curr))
#     print(head(df_curr))
    colnames(df_curr) = paste0(gsub("beta_GxE.tsv","",gene),"-",colnames(df_curr))
    df = cbind(df, df_curr)
}
nrow(df)
ncol(df)
head(df)
```
