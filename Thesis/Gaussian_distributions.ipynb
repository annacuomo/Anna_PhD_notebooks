```r
library(ggplot2)
```

```r
options(repr.plot.width = 4, repr.plot.height = 3)
ggplot(data.frame(x = c(-2, 2)), aes(x = x), n = 200) +
        stat_function(fun = dnorm) + theme_classic()
```
